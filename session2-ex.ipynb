1. We will first learn how to open datafiles in a csv (i.e. comma separated values) format. They are read into dataframes in Python which have a table-like structure.

```python
# opening a csv datafile into a dataframe
tweets_df = pd.read_csv('tweets.csv')

# (n is the number of rows you want it to show)
display(tweets_df.head(n = 20))

# structure of a dataframe (prints the number of rows and number of columns of the data frame)
tweets_df.shape
```

```python
# We can also do this sort of transformations on categorical variables. For example, lets create a variable called 
# 'is_apple'"tweets raw.csv" based on the 'source' variable. This variable will have a 'True' value if the source 
# of the post is an Apple device and a 'False' value otherwise. 

apple_list = ['Twitter for iPhone','Twitter for iPad','Tweetbot for iŒüS','Twitterrific for iOS','Twitter for Mac','OS X']
tweets_df['is_apple'] = [False for i in range(tweets_df.shape[0])]
for i in range(tweets_df.shape[0]):
    if tweets_df.loc[i,'source'] in apple_list:
        tweets_df.loc[i,'is_apple'] = True
    
# We can see that more than half of the posts were published from an Apple device
tweets_df['is_apple'].value_counts()
```

1. Often we want to see how many times each value of a variable appeared. When we are dealing with a nominal variable, we often use frequency tables. 

```python
# To generate the frequency table in absolute terms for the variable 'is_retweet'
display(tweets_df['is_retweet'].value_counts())

# To generate a relative (proportion) frequency table
tweets_df['is_retweet'].value_counts(normalize = True)
```

2. One could also check the cumulative distribution, where we count how many times a particular value or any other smaller value appeared. This makes most sense for ordinal variables. 

```python
# We first calculate a frequency table like before
freqs = tweets_df['sentiment_score'].value_counts().sort_index()

# We then add, to frequency of each value, that of all previous values
for i in freqs.index[1::]:
    freqs[i] = freqs[i]+freqs[i-1]

freqs
```

3. We now dive into data visualization (i.e. graphs). First, we are going to create a histogram. This sort of plot shows the, for interval data, how frequent values are. 

```python
# We first make a simple histogram for the variable 'favorite_count'
plt.hist(tweets_df['favorite_count'],bins=3000)
plt.xlim(0,200)
```

```python
# Then we make two other histograms via another code to show you how to select bins. 
# Below Seaborn selects the number of bins automatically.
sns.set_context("notebook", font_scale = 1)
sns.histplot(data = tweets_df, x="friends_count")
```

```python
# Now we change the code to define the number of bins by ourselves. This looks more appealing.
sns.set_context("notebook", font_scale = 1)

sns.histplot(data = tweets_df, x="friends_count", bins=30)
```
