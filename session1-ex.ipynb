3. A part from looking up existing variables, we can also create new variables from existing ones. Here we see an example where we transform an interval variable as well as one where we transform a nominal one. 

```python
# The variable 'display_text_width' indicates the length of the post. We are going to create a new variable 
# called 'relative_length' which will show the length of a post as the percentage of the maximum length.

# We first find the maximum length of a post
max_length = max(tweets_df['display_text_width'])
print(max_length)

# We then create the new variable 
relative_length = tweets_df['display_text_width']/max_length

# We finally add it trelative_lengtho our data set
tweets_df['relative_length'] = relative_length

# As we can see we now have an extra column (i.e. variable) in our dataframe
tweets_df.shape
```

```python
# We can also do this sort of transformations on categorical variables. For example, lets create a variable called 
# 'is_apple'"tweets raw.csv" based on the 'source' variable. This variable will have a 'True' value if the source 
# of the post is an Apple device and a 'False' value otherwise. 

apple_list = ['Twitter for iPhone','Twitter for iPad','Tweetbot for iŒüS','Twitterrific for iOS','Twitter for Mac','OS X']
tweets_df['is_apple'] = [False for i in range(tweets_df.shape[0])]
for i in range(tweets_df.shape[0]):
    if tweets_df.loc[i,'source'] in apple_list:
        tweets_df.loc[i,'is_apple'] = True
    
# We can see that more than half of the posts were published from an Apple device
tweets_df['is_apple'].value_counts()
```

4. Missing values (a.k.a. NA values, or NaN values) are common in data. They represent cells where the data is, for whatever reason, not available (e.g. because it could not be measured, because for a given case measuring would make no sense). Dealing with missing data is a complicated topic and it will depend greatly on the particular case. In this example we simply show how to make sure that NA values are treated as such in a Python dataframe. 


```python
# Sometimes NA values are not defined as such. For instance, a dataset may have a 9999 value in those cells 
# where data is missing. It is important to rename this values so they are correctly interpreted as NA.  

# We can specify this when we read the data in the first place pd.read_csv(...,na_values=9999) or now by using the replace()
# method.

tweets_df['reply_to_status_id'] = tweets_df['reply_to_status_id'].replace(9999,np.NaN)

# lets see the changes
tweets_df['reply_to_status_id']

```
